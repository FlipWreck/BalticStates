```python
from matplotlib import pyplot as plt

import pandas as pd
```

```python
# Sukurkite Python sąrašą (list) iš Baltijos šalių pavadinimų
baltic_countries = ["Lietuva","Latvija","Estija"]
```

```python
# Sukurkite Python sąrašą (list) su Baltijos šalių sostinėmis
baltic_capitals = ["Vilnius","Ryga","Talinas"]
```

```python
# Sukurkite Python žodyną (dictionary) su Baltijos šalių gyventojų skaičiumi
df_baltic_population = {
    "Lietuva":2794885,
    "Latvija":1900449,
    "Estija":1329522
}
```

```python
# Sukurkite Python žodyną (dictionary) su Baltijos šalių sostinių gyventojų skaičiumi 
# (duomenis imkite iš Vikipedijos)
df_baltic_capital_population = {
    "Vilnius":544386,
    "Ryga":632614,
    "Talinas":426538
}
```

```python
# Pavaizduokite Baltijos šalių gyventojų skaičių stulpelinėjė diagramoje
baltic_country_list = df_baltic_population.keys()
baltic_population_list = df_baltic_population.values()

plt.figure(figsize=(12,6))

bar = plt.bar(baltic_country_list, baltic_population_list, color="purple")

plt.title("\nBaltijos saliu gyventoju skaicius, mln.", fontsize=22)
plt.ylabel("Gyventojai, mln", fontsize=18)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Bar labels
labels_population_in_mln = [f"{round(p / 10**6, 2)} mln." for p in baltic_population_list]
plt.bar_label(bar, labels=labels_population_in_mln, padding=-30, fontsize=18, color="white")

plt.grid(axis="y")
plt.show()
```

```python
# Pavaizduokite Baltijos šalių sostinių gyventojų skaičių stulpelinėjė diagramoje.
baltic_capital_country_list = df_baltic_capital_population.keys()
baltic_capital_population_list = df_baltic_capital_population.values()

plt.figure(figsize=(12,6))

bar = plt.bar(baltic_capital_country_list, baltic_capital_population_list, color="purple")

plt.title("\nBaltijos saliu sostiniu gyventoju skaicius", fontsize=22)
plt.ylabel("Gyventojai", fontsize=18)

plt.xticks(fontsize=12)

# yticks formatting
ytick_loc, _ = plt.yticks()
ytick_label = ["{:,}".format(tick) for tick in ytick_loc]
plt.yticks(ticks=ytick_loc, labels=ytick_label, fontsize=12)

# Bar labels
labels = ["{:,}".format(p) for p in baltic_capital_population_list]
plt.bar_label(bar, labels=labels, padding=-30, fontsize=18, color="white", fmt="%0.0f")

plt.grid(axis="y")
plt.show()
```

```python
# Pavaizduokite Baltijos šalių gyventojų proporcijas skritulinėje diagramoje
plt.figure(figsize = (12, 8))

plt.pie(baltic_population_list, labels = baltic_countries, autopct = "%1.0f%%", textprops = {"fontsize":16})

plt.title("\nBaltijos saliu gyventoju pasiskirstymas", fontsize = 22)
plt.show()
```

```python
# Suraskite Lietuvos, Latvijos ir Estijos gyventojų skaičiaus dinamiką nuo 2015 iki 2020. 
# Pavaizduokite gyventojų skaičiaus keitimą linijinėje diagramoje.
df_baltic_population = pd.read_csv("data/baltic_population.csv", index_col="Year")

fig, ax = plt.subplots(figsize = (10, 6))

df_baltic_population.plot(ax = ax, linewidth=3)

plt.title("\nBaltijos saliu gyventoju skaiciaus dinamika\n", fontsize = 22)
plt.xlabel("\nYear", fontsize = 18)
plt.ylabel("\nGyventojai, mln.", fontsize = 18)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(loc="upper right", fontsize = 14, bbox_to_anchor = (1.4, 1))


plt.grid()
plt.show()
```

```python
# Suraskite Lietuvos, Latvijos ir Estijos gyventojų skaičiaus dinamiką nuo 2015 iki 2020. 
# Pavaizduokite gyventojų skaičiaus dinamiką, panuduojant Grouped Bar Chart tipo grafiką
fig, ax = plt.subplots(figsize = (12, 6))
df_baltic_population.plot.bar(ax = ax)

plt.title("\nBaltijos saliu gyventoju skaiciaus dinamika, mln. \n", fontsize = 22)
plt.xlabel("Year", fontsize = 18)
plt.ylabel("\nGyventojai, mln.", fontsize = 18)

plt.legend(loc="upper right", fontsize = 18, bbox_to_anchor = (1.4, 1))

plt.grid()
plt.show()
```
